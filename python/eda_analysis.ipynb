```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/ola_rides_50k.csv", parse_dates=['datetime'])

completed = df[df["ride_status"]=="Completed"]

city_revenue = completed.groupby("city")["fare_amount"].sum()

sns.barplot(x=city_revenue.index, y=city_revenue.values)
plt.xticks(rotation=45)
plt.title("Revenue by City")
plt.show()
```
